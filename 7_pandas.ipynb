##DataFrame
A DataFrame is a two-dimensional, tabular data structure used in data analysis and manipulation, primarily in Python with the pandas library. Think of it as an Excel spreadsheet or a SQL table, where data is organized in rows and columns.

#Key Features
##1. Rows and Columns:
*   Rows represent individual records or observations.
*   Columns represent attributes or variables describing the data.

##2. Labeled Index:
*  Rows and columns have labels (like row indices and column names) for easy data referencing.

##3. Heterogeneous Data:
* Each column can have a different data type (e.g., integers, floats, strings)


##4. Operations:



*   You can perform various operations like filtering, sorting, grouping, and statistical calculations.



```python
array_2d = np.arange(0,100).reshape(10,10)
array_2d
```

Let's create two labels/indexes:
* for rows 'r1 to r10'
* for columns 'c1 to c10'

Let's start with a simple example, using **`arange()`** and **`reshape()`** together to create a 2D array (matrix).<br>

```python
ind = 'r1 r2 r3 r4 r5 r6 r7 r8 r9 r10'.split(); #['r1', 'r2', 'r3', 'r4', 'r5', 'r6', 'r7', 'r8', 'r9', 'r10']

col = 'c1 c2 c3 c4 c5 c6 c7 c8 c9 c10'.split(); #['c1', 'c2', 'c3', 'c4', 'c5', 'c6', 'c7', 'c8', 'c9', 'c10']
```

##We need three part to make database 1. Index 2. Column 3. Data

```python
#df = Data Frame
df = pd.DataFrame(data=array_2d,index=ind,columns=col)
df #select * from df
#when so many series(columb) shares a single index(row) that is data set
```
