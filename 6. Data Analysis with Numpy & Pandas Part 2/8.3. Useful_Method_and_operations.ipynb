#Useful methods and operations
#####unique(),nunique(),value_counts(), sort_values(), apply(), index. column

```python
import numpy as np
import pandas as pd
import warnings #It ignores warnings
warnings.filterwarnings('ignore')
```

```python
data_dic = {
    'col':[1,2,3,4,5],
    'col_2':[111,222,333,111,555],
    'col_3':['abc','def','ghi',np.nan,np.nan]
}
df = pd.DataFrame(data_dic)
df
```

#unique()
Find and returns all the unique values.
Lets see how it works on all the columns in our dataframe.

```python
df['col_2'].unique()
```

#nunique() / Number of unique value
Find returns "how many unique values exist".<br>
-> Notice the difference, for NaN, it count a missing value and returns "3" for col_3.

```python
df['col_2'].nunique()
```

#value_counts()
We want a table with all the values along with no. of times they appeared in our data, value_counts do the work here!<br>
->for NaN. it count a missing value, nothing in the output.

```python
df['col_2'].value_counts()
```

```python
df['col_2'].count() #Number of elemens
```
