<h2 style='color:purple' align='center'>Training And Testing Available Data</h2>

<p><b>We have a dataset containing prices of used BMW cars. We are going to analyze this dataset
and build a prediction function that can predict a price by taking mileage and age of the car
as input. We will use sklearn train_test_split method to split training and testing dataset</b></p>

```python
import pandas as pd
```

```python
df =  pd.read_csv("/content/drive/MyDrive/ml_data/carprices.csv")
df
```

```python
#Indepndent Variable at X
X = df[['Mileage','Age(yrs)']]
X
```

```python
#Depndent Variable at Y
y = df[['Sell Price($)']]
y
```

This will split the features (X) and labels (y) into four parts:
- X_train: Training features
- X_test: Testing features
- Y_train: Training labels
- Y_test: Testing labels
- test_size=0.2: 20% of the dataset is used for testing, and the rest (80%) is for training.
- random_state=1: Setting a random seed ensures the split is consistent and reproducible each time the code is run.

```python
from sklearn.model_selection import train_test_split  # Import the function to split datasets into training and testing subsets

# Split the dataset into training and testing sets
# X: Features (input data)
# y: Target variable (labels)
# test_size=0.2: 20% of the data is used for testing, and the remaining 80% for training
# random_state=1: Ensures reproducibility of the split (same split every time)

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=1)
#Here 4 data test and 16 data training outof 20 because we have test size 20%
#we use random_state to get a different pattern and also if someone checks my model he and I ge the same values
```
