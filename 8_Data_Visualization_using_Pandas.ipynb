## Pie Plot
We can create a pie plot with  `Series.plot.pie()` or `DataFrame.plot.pie()`. If your data includes any NaN, they will be automatically filled with 0. A ValueError will be raised if there are any negative values in your data.<br>
Let's try pie plot with df2.<br>

We need to specify a target column by the y argument. When y is specified, pie plot of selected column will be drawn.

```python
df2
```

```python
df2.plot.pie(
    y='X',                # Specifies the column 'X' from DataFrame df2 to be used for the pie chart.
    figsize=(12, 10),     # Sets the figure size to 12 inches wide and 10 inches tall.
    autopct='%.1f%%'      # Displays percentage values on the pie chart with two decimal places.
)
```

```python
fifa = pd.read_csv("/content/drive/MyDrive/ml_data/fifa_data.csv")
fifa
```

```python
pd.set_option('display.max_columns', None)
# Sets the pandas display option to show all columns in the DataFrame when displayed.
# 'None' means there is no limit, so all columns will be visible regardless of the number of columns in the DataFrame.

fifa
```

```python
fifa['Weight']
```

```python
def stripdata(x):
  if type(x) == str:
    return float(x[0:3])
  else:
    return x
```

```python
stripdata('685kg')
```

```python
fifa['Weight']=fifa['Weight'].apply(stripdata)
```

```python
fifa['Weight']
```

```python
fifa.info()
```

```python
light = fifa[fifa['Weight']<125]
```

```python
light = fifa[fifa['Weight']<125].count()
```

```python
light = fifa[fifa['Weight']<125].count()[0] #I want just the first one
light
```

```python
light = fifa[fifa.Weight < 125].count()[0]
light_medium = fifa[(fifa.Weight >=125) & (fifa.Weight<150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[fifa.Weight >= 200].count()[0]

weights = [light,light_medium, medium, medium_heavy, heavy]
weights
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,8))
label = ['under 125', '125-150', '150-175', '175-200', 'over 200']
explodelist = [.4,.2,0,0,.4]

plt.pie(
    weights,               # The data values to be plotted as slices of the pie chart.
    labels=label,          # The labels corresponding to each slice.
    explode=explodelist,   # A list that determines the fraction to offset each slice from the center (e.g., [0.1, 0, 0] to offset the first slice).
    autopct='%1.1f%%',     # Formats the percentage values displayed on the slices with one decimal place.
    pctdistance=0.8        # Adjusts the distance of the percentage text from the center of the pie.
)

plt.show()
```
